```python
import pandas as pd
```

```python
df = pd.read_csv('Heart.csv')
df.head()
```

```python

# Ensure that the data does not contain missing values
df.isnull().sum()
```

```python
# Adding a column showing heart diseases and normal by converting zero and one into text
df['HD'] = df['HeartDisease'].apply(lambda x: 'heart disease' if x > 0 else 'Normal' if x == 0 else '')
df.head()
```
